# Dataframe - most used data structure of pandas


```python
# We can create dataframe from 2d array , dict of python ,read from diff type of files
# like csv,execl, parquet, json

def printLine():
    print('-'*40)



import pandas as pd

# dictionary/map - All arrays must be of the same length, type can be different
pyhton_dict={
        "name":["A","B","C","D"],
        "sub":["py","java","js","math"]
}

data_frame=pd.DataFrame(pyhton_dict)
print(data_frame)

# We can export this data frame to diff type like csv, excel,parquet...etc
data_frame.to_csv("./generatedDataFrame.csv")
```

```python

# Load dataframe from diff file
# pd.read_xxx(filepath)  xxx-> type of file 

bank_data=pd.read_csv("./BankDataSet.csv")   #data is loaded from file to bank_data obj

bank_data.head()  # get top elements, by default 5. Pass .head(totalRowFromTop)

bank_data.tail()  # similar to head just gives rows from bottom

bank_data.sample(5) # Random rows are selected from dataset, by default is 1 row ,pass value to fetch as much as we want

# why random ? 
# dataset can be very large and analysing data best on first or last few record doesn't make sense(Similar can be put at top or bottom)

# Practical usecase - Generate random data but same random data should be generated 

bank_data.sample(5,random_state=3) # here state will make sure same data is given again and again 
# state can be any random int, in short consider our state number is used to create randomness 
# so if we pass same state same random data will be given

```

```python
# few useful attributes for dataframe and series

list(bank_data.index)   # return rows (useful only if custom labels are given)
print(list(bank_data.columns)) # return list of columns in dataframe
printLine()
print("Diamension : ",bank_data.shape)  # diamension of dataframe
printLine()
print("size : ",bank_data.size) # get total size of dataframe (elements count)
printLine()
# print(bank_data.describe()) # gives description about dataframe , count,agg ,mean, min,max
print(bank_data.info())  # gives info about dataframe , datatypes ,memory usage


```
