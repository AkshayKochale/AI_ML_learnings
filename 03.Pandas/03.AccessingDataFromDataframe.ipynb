```python
#  method/ ways to acess data from dataframe 

import pandas as pd

bank_Data=pd.read_csv("./BankDataSet.csv")
bank_Data.head()  # for testing (display first 5 rows)

```

```python
# loc (location) is used to find data (single or range) from dataframe 
# label based indexing (elements can be access using labels)

# dataframe.loc[r,c]
# r = row which is start:end (from which row to which row), logic is similar to python slicing [s,e,jump]
# c = either it can be single as 'colName' or multiple as ['col1','col2'],for range ['col1':'col5']
bank_Data.loc[0:5,'Surname':'Gender']  # remeber r and c are strict and case sensitive 

# Note : if we want specific rows/col we have to metion it using ',' and  for range we need to use ':'

# iloc - similar to loc just searching based on index
bank_Data.iloc[0:4,4:7]  # for rows logic is same but instead of label we can use index starting from 0 to  lastcolindex

```

```python
# .at is used to find out specific data from dataframe using label based indexing
# why .at if we have .loc ?
# .at is optimized for single search operation and more efficient 

bank_Data.at[3,'Surname']  # only single search , range not supported 

# .iat - similar to .at just serch based on index 

bank_Data.iat[3,3]
```

```python
# also we can access elements directly

bank_Data['Surname']  #returns all the data in passes column

# when to use?
# when we want all the data as it has no control on range for rows
```
