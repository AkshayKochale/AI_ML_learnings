```python

#  method/ ways to filter data from dataframe 

import pandas as pd

bank_Data=pd.read_csv("./BankDataSet.csv")
bank_Data.head()  # for testing (display first 5 rows)
```

```python
# filtering means getting only data based on some condition
# for this we must use .loc as it support lable based indexing (search column based on lable )

bank_Data.loc[bank_Data['CreditScore']>=850]  # pass condition expression inside loc 

# We can pass multiple condition and combine them with & and | operators

bank_Data.loc[(bank_Data['CreditScore']>=850) & (bank_Data['Surname'].str.startswith('Mi'))]  

# syntax ->  dataframe.loc[ (condition1) & (condition2) ]
```

```python
# We can directly use this condition on dataframe without loc 

bank_Data[(bank_Data['CreditScore']>=850) & (bank_Data['Surname'].str.startswith('Mi'))] 
```
