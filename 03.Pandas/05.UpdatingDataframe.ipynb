```python
#  method/ ways to updaing data in dataframe 

import pandas as pd

bank_Data=pd.read_csv("./BankDataSet.csv")
bank_Data.head(3)  # for testing (display first 5 rows)
```

```python
# All the accessing methods can be use to update dataframe

bank_Data.loc[0:3,'Surname']='Kochale'    # surname col is updated
bank_Data.head(5)

#similar way you can use .iloc() , .at , .iat, bank_Data[expression] to update data 

# Note : All the changes are happening in dataframe not on acutal file 
# To actually make changes in file you need to write .to_csv() , as it save file from dataset to file
# pass same name to save update same file

```

```python
# Example of how to use filter and updating together

bank_Data.loc[list(bank_Data[bank_Data['Surname']=='Kochale'].index),"Surname"]="Akshay"
bank_Data.head(5)
```

```python
# Transformation - > Suppose you want to apply some function on all rows of specific column

def updateRows(x):
        return x+50

# For example suppose i want balance of each customer increased by 50
bank_Data['Balance']=bank_Data['Balance'].apply(updateRows)     

# Note: apply will not directly modify dataframe, we must assign value to dataframe column

# Also we can use lamda if function are simple 
bank_Data['Balance']=bank_Data['Balance'].apply((lambda x:x+50))     
 

# Also we can use .apply to add new column in dataframe
def addCategory(x):
        if x['Balance']>=10000 and x['Age']>30:return "Rich" 
        else : return "Middle"

bank_Data['Category']=bank_Data.apply(addCategory,axis=1)
bank_Data.head(5)
```

```python

# .insert(index,name,value) can be also use to add column at specific index 
# We can pass expression at place of value to dynamically generate based on some previously present data in dataframe

bank_Data.insert(1,'FirstName','N/A')

# Also to drop column we can use .drop(colname,inplace)
bank_Data.drop(columns='FirstName',inplace=True)    # inplace make changes in dataframe
bank_Data.head(5)
```
