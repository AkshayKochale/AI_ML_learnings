    # 3D plot (rarely used but good to know)

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.random.rand(50)*10
y =np.random.rand(50)*10
z =np.random.rand(50)*10

plt.axes(projection='3d')  # set projection to 3d 

plt.scatter(x,y,z)
plt.show()

# Figure out as per you requirement from documentation of Matplotlib
```
