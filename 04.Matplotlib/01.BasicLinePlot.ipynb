```python

# Basic line chart

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.plot(x,y,label='Score1')   # pass value of x and y axis 
plt.title("Demo plot")  # set name for graph
plt.xlabel("Series")   # set name for x axis
plt.ylabel("Points")   # set name for y axis


# Set what values will be visible on x and y axis 
# here i am passing x and y list as i want same values on both axis 
# but values can be diff and we can set using xticks([]) or yticks([]) 
plt.xticks(x)
plt.yticks(y)

# Similary you can pass another values to draw 2 or more lines
z=[15,20,33,25,20]
plt.plot(x,z,label='Score2')

# To set legends (Name for each line)
plt.legend(loc="upper left")

# Save plot in local machine -> diff formats are supported like png,jpeg,pdf,svg etc.
plt.savefig("MyDemoPlot.png")
# Note : plot should be save before .show(), as matlplotlib as single instace of each graph 
# and when show is called grpah is rendered and we dont have access to graph obj any more


# To display actual chart (optional in Jupyter)
plt.show() 


# In addition to this there is option of making grpah more readable 
# like change font ,add markup (symbols at point), font size 
# but those are optional, read more on https://matplotlib.org/stable/users/explain/quick_start.html 
```
