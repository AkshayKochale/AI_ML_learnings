```python
import matplotlib.pyplot as plt
import numpy as np

# 01. Scatter graph 

x= np.random.rand(50)*10
y= np.random.rand(50)*10
z= np.random.rand(50)*10


plt.scatter(x,y)
plt.scatter(x,z)

# All the logic is same as normal plot we can customize look of graph, Also plot mutiple on same graph

```

```python


# 02. Bar graph

x=['Stu1','Stu2','Stu3','Stu4']
y=[60,70,90,50]

plt.bar(x,y)
plt.show()

# Same customization is applied as simple line plot
```

```python

#03. Histogram

# what is diff between bar graph and histogram?
# Even both look similar but bar graphs takes (x,y) which is what is value of y for each x
# and histogram takes only (x) and  shows what is frequecy of each numerical value in dataset

# Histograms each category (unique number ) is reffered as bin
# we can customzied how much bean we want

flist=[3,4,5,5,2,2,3,4,4,6,2,7,1]
plt.hist(flist,bins=7)
plt.xlabel("Numbers")
plt.ylabel("Frequency in list")
plt.show()
```

```python
# 04. Pie chart

#show pie part outside, called as explode
explode_prop=[0.2,0,0,0]
plt.pie(y,labels=x,explode=explode_prop,shadow=True,autopct="%2.f%%")   # here data on y-axis is plotted with x as lables
# autopct to add percentage (useful)

plt.show()
```
