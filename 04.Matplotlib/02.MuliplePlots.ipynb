```python

# Muliple graph to render either horizontally or vertically
# Previously we see that we can plot muliple lines in same grpah, but we want graph for each plot parameter

import matplotlib.pyplot as plt


# some random value 
# x is for x-axis and  p,q,r for y-axis (mulitple lines)
x=[10,20,30,40,50]   
p=[25,35,20,10,35]
q=[90,50,30,35,20]
r=[50,80,30,50,70]

# To plot all valeus at same graph we can use diff syntax

plt.plot(x,p,x,q,x,r)   # instead of writing 3 diff plot we can pass multiple (x,y,x,y,x,y)
plt.show()

```

```python

# To create 3 diff graphs we must use 'figure' (each graph is figure in matplotlib)
# and create subplot for each line, so we can get 3 graph

plt.figure()   # internally one single figure is created 

# Suppose we want 3 graphs horizontally 
# then we can create 3 subplot and mentions at which index we want out current plot
# here .subplot(1,3,1) means 1 row , 3 cols ,1st index (similar to matrix just row and col is repeated in each subplot)
# .subplot(r,c,i)

plt.subplot(1,3,1)
# .. do some formating if required
plt.plot(x,p)

plt.subplot(1,3,2)
# .. do some formating if required
plt.plot(x,q)

plt.subplot(1,3,3)
# .. do some formating if required
plt.plot(x,r)

plt.show()

# Similarly by changing plot r,c we can change orientaion from horiontal to vertical
# like .subplot(3,1,1)   -> total 3 rows , 1 col and at index 1

# Note : Formatting logic is same as normal plots.
```
